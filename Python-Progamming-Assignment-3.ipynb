```python
a=int(input("Please enter a natural no"))
if a<0:
    print("It is negative no")
elif a>0:
    print("It is positive no")
elif a==0:
    print("It is zero")
```

```python
b=int(input("Please enter a natural no"))
if b<0:
    print("Enter a positive no")
elif b==0:
    print("0 is neither odd or even")
else :
    if b%2==0:
        print(b,"is a even no ")
    if b%2==1:
        print(b,"is a odd no ")
    
    
```

```python
c=int(input("Enter a year"))
if c<0:
    print("Enter a valid year")
else :
    if c%4==0:
        if c%400==0:
            print("It is a leap year")
        elif c%100==0:
            print("It is not leap year")
        else:
            print("It is a leap year")
    else:
        print("Not a leap year")
            
            
```

```python
d=int(input("Enter a no to check whether it is prime number or not "))

for i in range(2,d):
    if d%i==0:
        print("The no is composite")
        break
        
else:
    print("The no is prime")
```
